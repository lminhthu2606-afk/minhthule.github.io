```
# [cite_start]5. The Gatekeeper (Function) [cite: 65]
print("\n### 5. The Gatekeeper (Function) ###")

# 1. Function identify_type(x)
def identify_type(x):
    return type(x)

# 2. Function identify_types(*args)
def identify_types(*args):
    return [type(x) for x in args] # S·ª≠ d·ª•ng list comprehension

# Test Step 2
test_types = identify_types(10, 'magic')
[cite_start]print(f"‚Ä¢ identify_types (10, 'magic') ‚Üí {[t.__name__ for t in test_types]}") # Expected: [int, str] [cite: 74]

# 3. Function process_list(lst) using for loop
def process_list_for(lst):
    result = []
    for item in lst:
        if isinstance(item, str):
            result.append(0)
        elif isinstance(item, (int, float)):
            result.append(item)
    return result

# 4. Rewrite step 3 using while loop
def process_list_while(lst):
    result = []
    index = 0
    while index < len(lst):
        item = lst[index]
        if isinstance(item, str):
            result.append(0)
        elif isinstance(item, (int, float)):
            result.append(item)
        index += 1
    return result

# Test Step 3/4
test_list = [10, 'magic', 20]
[cite_start]print(f"‚Ä¢ process_list (for): {process_list_for(test_list)}") # Expected: [10, 0, 20] [cite: 76]
print(f"‚Ä¢ process_list (while): {process_list_while(test_list)}")
```

### 7\. The Selective Gate (List & Conditions)

**Logic:**

1.  [cite_start]ƒê·ªãnh nghƒ©a h√†m `filter_numbers(lst)`. [cite: 89]
2.  [cite_start]D√πng list comprehension ƒë·ªÉ l·ªçc: ch·ªâ gi·ªØ l·∫°i c√°c s·ªë th·ªèa m√£n `num % 2 == 0` (chia h·∫øt cho 2) **v√†** `num % 3 != 0` (kh√¥ng chia h·∫øt cho 3). [cite: 90]
3.  [cite_start]S·ª≠ d·ª•ng `len()` ƒë·ªÉ ƒë·∫øm s·ªë l∆∞·ª£ng v√† `sum()` ƒë·ªÉ t√≠nh t·ªïng c·ªßa list ƒë√£ l·ªçc. [cite: 92, 93]

**M√£ ngu·ªìn Python:**

```
# [cite_start]7. The Selective Gate (List & Conditions) [cite: 87]
print("\n### 7. The Selective Gate ###")
def filter_numbers(lst):
    # Filter numbers: divisible by 2 AND NOT divisible by 3
    filtered = [
        num for num in lst
        if num % 2 == 0 and num % 3 != 0
    ]

    count = len(filtered)
    sum_val = sum(filtered)

    return count, sum_val

# Test Case
test_list = [2, 3, 4, 6, 8, 9, 10]
count, sum_val = filter_numbers(test_list)
print(f"‚Ä¢ Input: {test_list}")
[cite_start]print(f"‚Ä¢ Expected Output: Count: {count}, Sum: {sum_val}") # Expected: Count: 4, Sum: 24 [cite: 97]
```

### 6\. The Clusters (kMeans Logic)

**Logic:**

1.  [cite_start]**H√†m kho·∫£ng c√°ch Euclidean:** √Åp d·ª•ng c√¥ng th·ª©c $d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$. [cite: 427, 428]
2.  [cite_start]T√≠nh kho·∫£ng c√°ch t·ª´ ƒëi·ªÉm $A$ ƒë·∫øn $C_1$ v√† $C_2$. [cite: 427, 428]
3.  [cite_start]So s√°nh hai kho·∫£ng c√°ch v√† g√°n $A$ v√†o Centroid g·∫ßn h∆°n (kho·∫£ng c√°ch nh·ªè h∆°n). [cite: 429]

**M√£ ngu·ªìn Python:**

```
import math

# [cite_start]6. The Clusters (kMeans Logic) [cite: 425]
print("\n### 6. The Clusters (kMeans Logic) ###")

def euclidean_distance_points(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def kmeans_assignment(A, C1, C2):
    # [cite_start]1. Dist A to C1 [cite: 427]
    dist_A_C1 = euclidean_distance_points(A, C1)

    # [cite_start]2. Dist A to C2 [cite: 428]
    dist_A_C2 = euclidean_distance_points(A, C2)

    # [cite_start]3. Assignment [cite: 429]
    if dist_A_C1 <= dist_A_C2:
        nearest_centroid = "C1"
        nearest_distance = dist_A_C1
    else:
        nearest_centroid = "C2"
        nearest_distance = dist_A_C2

    return nearest_distance, nearest_centroid

# Test Case
A_point = (1, 1)
C1_point = (1, 0)
[cite_start]C2_point = (10, 11) # T·ª´ Context [cite: 426]

nearest_dist, assignment = kmeans_assignment(A_point, C1_point, C2_point)

[cite_start]print(f"‚Ä¢ Input: A={A_point}, C1={C1_point}") # [cite: 431]
print(f"‚Ä¢ Kho·∫£ng c√°ch g·∫ßn nh·∫•t: {nearest_dist:.1f}. G√°n cho: {assignment}.") # Expected: Dist=1.0. [cite_start]Assigned to C1. [cite: 432]
```

### 7\. The Unknown Plant (kNN Logic)

**Logic:**

1.  [cite_start]T√≠nh kho·∫£ng c√°ch (tuy·ªát ƒë·ªëi) t·ª´ ƒëi·ªÉm m·ª•c ti√™u $Y$ ƒë·∫øn m·ªói ƒëi·ªÉm $X_i$ trong t·∫≠p d·ªØ li·ªáu: `distance = abs(Y - X_i)`. [cite: 435]
2.  S·∫Øp x·∫øp c√°c ƒëi·ªÉm theo kho·∫£ng c√°ch tƒÉng d·∫ßn.
3.  [cite_start]L·∫•y ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n (nearest neighbor - h√†ng x√≥m g·∫ßn nh·∫•t) v√† tr·∫£ v·ªÅ nh√£n c·ªßa n√≥. [cite: 436]

**M√£ ngu·ªìn Python:**

```
# [cite_start]7. The Unknown Plant (kNN Logic) [cite: 433]
print("\n### 7. The Unknown Plant (kNN Logic) ###")

def knn_logic(X, Labels, Y, k=1):

    # 1. Calculate absolute difference (distance)
    distances = []
    for i, x_point in enumerate(X):
        [cite_start]distance = abs(Y - x_point) # [cite: 435]
        distances.append({
            'label': Labels[i],
            'distance': distance
        })

    # S·∫Øp x·∫øp theo kho·∫£ng c√°ch
    distances.sort(key=lambda item: item['distance'])

    # 2. Find the 1 nearest neighbor
    [cite_start]nearest_label = distances[0]['label'] # [cite: 436]

    return nearest_label

# Test Case
X_data = [1, 2, 10]
Labels_data = ['Small', 'Small', 'Big']
[cite_start]Y_target = 3 # [cite: 438]

result_label = knn_logic(X_data, Labels_data, Y_target, k=1)

print(f"‚Ä¢ Input: Y={Y_target}")
[cite_start]print(f"‚Ä¢ Nh√£n d·ª± ƒëo√°n: '{result_label}' (H√†ng x√≥m g·∫ßn nh·∫•t l√† 2)") # Expected: Label: 'Small' (Nearest neighbor is 2) [cite: 439]
```

```
# [cite_start]2. Conan nh√† t·∫≠p [cite: 583]
print("\n### 2. Conan nh√† t·∫≠p (Metrics) ###")

def calculate_metrics(y_true, y_pred):
    [cite_start]TP, TN, FP, FN = 0, 0, 0, 0 # [cite: 589]
    for true, pred in zip(y_true, y_pred):
        if true == 1 and pred == 1: TP += 1
        elif true == 0 and pred == 0: TN += 1
        elif true == 0 and pred == 1: FP += 1
        elif true == 1 and pred == 0: FN += 1

    [cite_start]precision = TP / (TP + FP) if (TP + FP) > 0 else 0 # [cite: 590]
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0

    return TP, TN, FP, FN, precision, recall

[cite_start]y_true_input = [0, 0, 1, 1, 0, 1, 1, 0] # [cite: 586]
[cite_start]y_pred_input = [0, 1, 1, 0, 0, 1, 1, 1] # [cite: 587]

# 1. Th·ªëng k√™ chung
total = len(y_true_input)
actual_thieves = sum(y_true_input)
suspects = sum(y_pred_input)
[cite_start]print(f"‚Ä¢ T·ªïng: {total} ng∆∞·ªùi, Th·ª±c t·∫ø tr·ªôm: {actual_thieves}, B·ªã nghi ng·ªù: {suspects}") # [cite: 597]

# 2 & 3. Confusion Matrix & Metrics
TP, TN, FP, FN, P, R = calculate_metrics(y_true_input, y_pred_input)
[cite_start]print(f"‚Ä¢ TP={TP}, TN={TN}, FP={FP}, FN={FN}") # [cite: 598]
[cite_start]print(f"‚Ä¢ Precision = {P:.2f}, Recall = {R:.2f}") # [cite: 599]
[cite_start]print("  Ch·ªâ s·ªë **Recall quan tr·ªçng h∆°n** v√¨ kh√¥ng mu·ªën b·ªè s√≥t k·∫ª tr·ªôm (gi·∫£m FN). [cite: 591]")

# 4. M√¥ ph·ªèng sau ƒëi·ªÅu ch·ªânh
[cite_start]y_pred_new = [1, 1, 1, 1, 0, 1, 1, 1] # [cite: 593]
TP_new, TN_new, FP_new, FN_new, P_new, R_new = calculate_metrics(y_true_input, y_pred_new)
[cite_start]print(f"‚Ä¢ Sau ƒëi·ªÅu ch·ªânh: Precision = {P_new:.2f}, Recall = {R_new:.1f} (trade-off)") # [cite: 599]
```
